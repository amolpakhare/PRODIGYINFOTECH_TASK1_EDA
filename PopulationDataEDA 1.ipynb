## Extraction of top 10 country with respect to total population 

```python
Total_Pop_data = pop_df[pop_df['Series Code']== 'SP.POP.TOTL.FE.IN']
Total_Pop_sorted = Total_Pop_data.sort_values(by='2017',ascending=False)
Total_toptencountries = Total_Pop_sorted.head(10)
Total_toptencountries
```

## Plot Bar chart top 10 countries

```python


plt.figure(figsize=(10, 6))
sns.barplot(x='2020',y='Country Code',data=Total_toptencountries)
plt.title('Bar Plot Example')
plt.xlabel('Category')
plt.ylabel('Values')
plt.show()
```

Conclusion: In 2018 China is 1st no in population then India 2nd and 3rd USA

## Extraction of Bottom 10 country with respect to total population

```python
Total_Pop_data = pop_df[pop_df['Series Code']== 'SP.POP.TOTL.FE.IN']
Total_Pop_sorted = Total_Pop_data.sort_values(by='2017',ascending=True)
Total_Bottomtencountries = Total_Pop_sorted.head(10)
Total_Bottomtencountries
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='2018',y='Country Code',data=Total_Bottomtencountries)
plt.title('Bar Plot Example')
plt.xlabel('Category')
plt.ylabel('Values')
plt.show()

plt.figure(figsize=(10, 6))
sns.barplot(x='2020',y='Country Code',data=Total_Bottomtencountries)
plt.title('Bar Plot Example')
plt.xlabel('Category')
plt.ylabel('Values')
plt.show()
```

Conclusion: The bottom country in 2018 is Tuvalu
